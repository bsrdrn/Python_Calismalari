**Listelerin Fonksiyonları**
* append : Listenin sonuna eleman eklemek için kullanılır
* extend : Listeleri birleştirmek için kullanılır. Birleştirmek için extend fonksiyonu gibi + operatörü de kullanılabilir
* insert : Listenin belirtilen konumuna eleman eklemek için kullanılır. 
* remove : Listenin içindeki değeri verilen elemanı siler. Herhangi bir eleman indis numarasına göre de silinebilir
* pop : Listede belirtilen konumdaki (indexteki) elemanı siler.
* clear : Listenin tüm elemanlarını siler.
* index : Bir elemanın listedeki konumunu buldurur.
* count : İstenen elemandan listede kaç tane olduğunu saydırır.
* sort :  Listenin içindeki elemanları string, int vb. veri tiplerine uygun olarak sıralar.
* reverse :Listeyi sondan başlıycak şekilde ters yazdırır.
* copy : Listeyi yeni bir liste olarak kopyalar.
* del : indeksi verilen elemanı siler. Pop fonksiyonuna benzer bir fonksiyon olmasına rağmen kullanımı farklıdır.

```python
numbers = [1, 3, 4, 16, 1, 8, 5, 9]
letter = ["a", "b", "y", "g", "z", "a","s","y"]

val = min(numbers)
val1 = max(numbers)
val2 = min(letter)
val3 = max(letter)

print(val)
print(val1)
print(val2)
print(val3)

numbers[4]=40

print(numbers)

numbers.append(49)     
numbers.insert(3,78)    
numbers.insert(-1,23)
print(numbers)

numbers.pop()           
numbers.pop(0)          
print(numbers)          

numbers.remove(16)      
print(numbers)

numbers.sort()
letter.sort()

print(numbers)
print(letter)

numbers.reverse()
letter.reverse()

print(numbers)
print(letter)

letter.clear()
print(letter)
```

```python
from unicodedata import name


names = ["Ali", "yağmur", "hakan", "deniz"]
years = [1998, 2000, 1998, 1987]

#"Cenk" ismini listenin sonuna ekleyiniz.
names.append("Cenk")
print(names)


#"Sena" ismini listenin başına ekleyiniz.
names.insert(0,"Sena")
print(names)


#"deniz" ismini listeden siliniz.
names.remove("deniz")
print(names)


#"deniz" isminin indexi nedir?
index = names.index("hakan")
print(index)

#"Ali" listenin bir elemanı mıdır?
result="Ali" in names
print(result)


#Liste elemanlarını ters çevirin.
print(names[::-1])


#Liste elemanlarını alfabetik olarak sıralayın.
names.sort()
print(names)


#Years listesini rakamsal büyüklüğüne göre sıralayınız.
years.sort()
print(years)


#str = "Chevrolet, Dacia" karakter dizisini listeye çevirin.
str = "Chevrolet, Dacia"
result = str.split(",")
print(result)


#Years dizisinin en büyük ve en küçük karakteri?
buyuk = max(years)
kucuk = min(years)

print(buyuk, kucuk)


#Years dizisinde kaç tane 1998 değeri vardır?
result = years.count(1998)
print(result)


#Years dizisinden tüm elemanları silin.
years.clear()
print(years)


#Kullanıcıdan alacağınız 3 marka bilgisini listede saklayın.
```
